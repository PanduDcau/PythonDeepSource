### Data Acquisition 

```python
import pandas as pd
import numpy as np
```

```python
# First 5 rows
df = pd.read_csv("automobile.csv")
df.head()
```

```python
# Last 5 rows
df.tail()
```

```python
# Count missing data
missing_data = df.isnull()
missing_data.head(5)
```

```python
# Create headers
#headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         #"drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         #"num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
        # "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
# Apply column names to headers
#df.columns = headers
df.head()
```

```python
# Replace (?) with (np.NaN)
df1 = df.replace("?", 0)
df1.head(5)
```

```python
# drop missing values
df = df1.dropna(subset = ["price"], axis = 0)
df.head()
```

```python
# Name of columns
df.columns
```

```python
# Save DataFrame
df.to_csv("automobile.csv", index = False)
```

```python
# Replece missing values with avg 
df["stroke"] = df["stroke"].astype("float")
mean_stroke = df["stroke"].mean(axis = 0)
df["stroke"].replace(np.nan, mean_stroke, inplace = True)
df.head()
```

```python
df["normalized-losses"] = df["normalized-losses"].astype("int")
```

```python
# replace missing values of (normalized-losses) with mean values
mean_norm = df["normalized-losses"].mean()
df["normalized-losses"].replace(np.nan, mean_norm)
df.describe()
```

```python
# Convert mpg to L/100km
df["city-mpg"] = 235 / df["city-mpg"]
df.head()
```

```python
df.rename(columns = {"city-mpg": "city-L/100km"}, inplace = True)
```

```python
# simple feature scaling in python
df["length"] = df["length"] / df["length"].max()
df.head()
```

```python
# Min - Max in python
df["length"] = (df["length"] - df["length"].min()) / (df["length"].max() - df["length"].min())
df.head()
```

### Z - Score

```python
df["length"] = (df["length"] - df["length"].mean()) / df["length"].std()
df.head()
```

```python
df["price"] = df["price"].astype("int")
```

### Binning

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

### Dummies

```python
dummy = pd.get_dummies(df["aspiration"])
dummy.rename(columns = {"std": "std aspiration", "turbo": "turbo aspiration"}, inplace = True)
df = pd.concat([df, dummy], axis=1)
df.drop('aspiration', axis = 1, inplace=True)
```
