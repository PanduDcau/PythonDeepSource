Line Chart for Income and Expenses by Month

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
income = np.array([260000, 278000, 284000, 290000, 295000, 300000, 308000, 311000, 318000, 320000, 325000, 332000])
expenses = np.array([198000, 195000, 202000, 210000, 206000, 212000, 215000, 214000, 218000, 220000, 225000, 226000])

# Plotting the line chart
plt.plot(months, income, label='Income')
plt.plot(months, expenses, label='Expenses')
plt.title("Income and Expenses by Months")
plt.xlabel('Months')
plt.ylabel('Amount (in Rs)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

Column Chart (Bar Chart) to Visualize Months and Income

```python
# Plotting the bar chart
plt.bar(months, income, color='blue')
plt.title("Income by Months")
plt.xlabel('Months')
plt.ylabel('Income (in Rs)')
plt.show()
```

Horizontal Bar Chart to Visualize Months and Income

```python
# Plotting the horizontal bar chart
plt.barh(months, income, color='green')
plt.title("Income by Months")
plt.xlabel('Income (in Rs)')
plt.ylabel('Months')
plt.show()
```

4. Horizontal Stacked Bar Chart to Visualize Income and Expenses by Months

```python
# Plotting the stacked horizontal bar chart
plt.barh(months, income, color='blue', label='Income')
plt.barh(months, expenses, color='orange', label='Expenses', left=income-expenses)
plt.title("Income and Expenses by Months")
plt.xlabel('Amount (in Rs)')
plt.ylabel('Months')
plt.legend()
plt.show()
```

5. Stacked Column Chart to Visualize Income and Expenses by Months

```python
# Plotting the stacked column chart
plt.bar(months, income, color='blue', label='Income')
plt.bar(months, expenses, color='orange', label='Expenses', bottom=income-expenses)
plt.title("Income and Expenses by Months")
plt.xlabel('Months')
plt.ylabel('Amount (in Rs)')
plt.legend()
plt.show()
```

6. Histogram of 100 Random Integer Values

```python
# Generating 100 random integer values and plotting a histogram
random_values = np.random.randint(0, 101, 100)
plt.hist(random_values, bins=10, color='purple')
plt.title("Histogram of Random Integer Values")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

7. Pie Chart for Vehicle Type Counts

```python
import pandas as pd

# Data for the pie chart
vehicle_data = {
    'Vehicle Type': ['Motorcycle', 'Car', 'Van', 'Lorry', 'Bus'],
    'Count': [146, 85, 32, 43, 65]
}

df_vehicle = pd.DataFrame(vehicle_data)

# Plotting the pie chart
plt.pie(df_vehicle['Count'], labels=df_vehicle['Vehicle Type'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Vehicle Types')
plt.show()
```
