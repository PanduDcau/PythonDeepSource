### Linear regression

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
model = LinearRegression(fit_intercept = True)
model
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data
```

```python
featuresAll=[]
features = iris.data[: , [0,1,2,3]]
```

```python
for observation in features:
    featuresAll.append([observation[0] + observation[1] + observation[2] + observation[3]])
print(featuresAll)
```

```python
#Finding the relationship between Sepal Length and Sepal width
featuresAll = []
targets = []
for feature in features:
    featuresAll.append(feature[0]) #Sepal length
    targets.append(feature[1]) #sepal width
```

```python
XX = np.asarray(featuresAll)
X2 = XX[:, np.newaxis]
X2.shape
```

```python
y2 = iris.target
y2.shape
```

```python
model.fit(X2, y2)
```

```python
# coefficient
model.coef_
```

```python
# intercept
model.intercept_
```

```python
Xfit = np.random.randint(8, size = (150))
Xfit.astype(float)
Xfit = Xfit[:, np.newaxis]
Xfit.shape
```

```python
yfit = model.predict(Xfit)
yfit.shape
```

```python
plt.scatter(X2, y2)
plt.plot(Xfit, yfit)
```

```python
# Polynomial Features

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(150, include_bias = False)
poly.fit_transform(X2)
```

```python
from sklearn.pipeline import make_pipeline
poly_model = make_pipeline(PolynomialFeatures(3), LinearRegression())
poly_model.fit(X2, y2)
yfit = poly_model.predict(Xfit)
```

```python
plt.scatter(X2, y2)
plt.plot(Xfit, yfit)
```

**How the length and width vary according to the species**

```python
iris = pd.read_csv("Iris.csv")
iris.head(5)
```

**Scatter Plot with Iris Dataset (Relationship between Sepal Length and SepalWidth)**

```python
iris.plot(kind = "scatter", x = "SepalLengthCm", y = "SepalWidthCm")
plt.show()
```

**Scatter Plot with Iris Dataset (Relationship between Petal Length and Petal Width)**

```python
iris.plot(kind = "scatter", x = "PetalLengthCm", y = "PetalWidthCm")
plt.show()
```

**Violin Plot**

```python
import seaborn as sns
plt.figure(figsize = (15, 10))
plt.subplot(2, 2, 1)
sns.violinplot(x = "Species", y = "PetalLengthCm", data = iris)
plt.subplot(2, 2, 2)
sns.violinplot(x = "Species", y = "PetalWidthCm", data = iris)
plt.subplot(2, 2, 3)
sns.violinplot(x = "Species", y = "SepalLengthCm", data = iris)
plt.subplot(2, 2, 4)
sns.violinplot(x = "Species", y = "SepalWidthCm", data = iris)
```
